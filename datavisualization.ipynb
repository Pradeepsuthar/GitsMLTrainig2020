```
# Data Visualization
# in systems : pip install matplotlib seaborn
# in colab : !pip install matplotlib seaborn

import matplotlib.pyplot as plt

plt.xlabel('Distance') # for labeling x axis
plt.ylabel('Time') # for y 
x = [2,5,4,9]
y = [3,8,3,6]
plt.scatter(x,y,label="minsss") # to show dots only
plt.plot(x,y,label="railway track") # to plot line
plt.bar(x,y,color='green',label="finace growth") # to show bar plot
plt.bar(y,x,color='orange')
plt.grid(c='green') # for grid and color as well
plt.legend()
plt.plot() # to plot the graph

```

```
# Criket score graph
players=["dhoni", "rohi","virat"]
runs=[120,160,130]

# plt.xlabel("Players")
# plt.ylabel("Runs")

# plt.bar(players,runs,label="202 starting match")
plt.pie(runs,labels=players,autopct='%1.1f%%',explode=(0,0.1,0),shadow=True)
plt.legend()
plt.show()
```
